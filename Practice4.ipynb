```python
path = kagglehub.dataset_download("peterkmutua/student-hours-scores")
```

```python
import os
print(os.listdir(path))
csv_path=path+"/Hours and Scores.csv"
print(csv_path)
```

```python
import pandas as pd

df = pd.read_csv(csv_path)

print(df.head())
```

```python
print(df.shape)
print(df.columns)
```

```python
print(df.isnull().sum())
```

```python
print(df.fillna(0))
```

```python
X=df[["Hours"]]
Y=df["Scores"]
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
print(X_train)
print(X_test)

```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,Y_train)
print("Model trained successfully")
```

```python
prediction=model.predict(X_test)
print(prediction)
```

```python
new_prediction=model.predict([[9]])
print("Prediction by ml model is:",new_prediction)
```
