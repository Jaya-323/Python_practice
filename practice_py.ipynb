Average marks finder


```python
def average(a,b,c):
  return (a+b+c)/3

f=float(input("Enter first number:"))
g=float(input("Enter second number:"))
h=float(input("Enter third number:"))
result=average(f,g,h)
print("The average is",round(result,2))
```

Grade finder for students

Print statement is written outside while calling function only when there is return keyword used in function instead of print function itself


```python
def grade(n):

  if(n>90):
     print("Grade A+")
  elif(n>80 and n<=90):
     print("Grade A")
  elif(n>70 and n<=80):
     print("Grade B")
  elif(n>60 and n<=70):
     print("Grade c")
  elif(n>50 and n<=60):
     print("Grade D")
  elif(n>40 and n<=50):
     print("Grade E")
  elif(n>30 and n<=40):
     print("Grade F")
  else:
     print("Fail")


def average(sum,d):
  return sum/d

n=int(input("Enter number of subjects:"))
s=0
for i in range(n):
   a=float(input("Enter your marks:"))
   s+=a;

result=average(s,n)
print("The average is",round(result,2))

p=grade(result)

```
