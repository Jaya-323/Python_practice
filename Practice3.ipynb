Finding average marks from a list of marks

```python
def avg(marks):
   return sum(marks)/len(marks)

list=[98,89,97,95,81,81.5,85,5]
print(f"Average marks is:{avg(list)}")
```

Grade checker

```python
def grade():
   n=int(input("Enter marks:"))
   if(n>=90 and n<=100):
      print("Grade A")
   elif(n>=80 and n<90):
      print("Grade B")
   elif(n>=70 and n<80):
      print("Grade C")
   else:
     print("Failed")

grade()

```

Exception handling using try and except

```python
try:
   n=int(input("Enter a number"))
   m=int(input("Enter a number"))
   print(n/m)
except:
   if(m==0):
      print("Division by 0 is not possible")
```
